```python
idades = []

with open(file='./banco.csv', mode='r', encoding='utf8') as arquivo:
  linha = arquivo.readline()
  linha = arquivo.readline()
  while linha:
    linha_separada = linha.split(sep=',')
    idade = linha_separada[0]
    idade = int(idade)
    idades.append(idade)
    linha = arquivo.readline()

print(idades)
```

```python
with open(file='idades.csv', mode='w',encoding='utf8') as fp:
  linha = 'idade' + '\n'
  fp.write(linha)
  for idade in idades:
    linha = str(idade) + '\n'
    fp.write(linha)
```
